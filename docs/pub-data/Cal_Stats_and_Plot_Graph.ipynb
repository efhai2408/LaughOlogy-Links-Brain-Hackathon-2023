```python
import pandas as pd
import numpy as np

# LINK = 'https://raw.githubusercontent.com/tangnatta/LaughOlogy-Links-Brain-Hackathon-2023/pub-data/pub-data/num_nodes1.csv'
# LINK = 'https://raw.githubusercontent.com/tangnatta/LaughOlogy-Links-Brain-Hackathon-2023/pub-data/pub-data/twodata3.csv'
LINK = 'https://raw.githubusercontent.com/tangnatta/LaughOlogy-Links-Brain-Hackathon-2023/pub-data/pub-data/twodata4.csv'

df = pd.read_csv(LINK)
```

```python
# TEST_COLUMNS = ["num_nodes_joke", "num_nodes_nonj"]
# TEST_COLUMNS = ["num_edges_joke", "num_edges_nonj"]
# TEST_COLUMNS = ["density_joke", "density_nonj"]
TEST_COLUMNS = ["deg_assortativity_coefficient_joke", "deg_assortativity_coefficient_njok"]
# TEST_COLUMNS = ["mean_clustering_joke", "mean_clustering_nonj"]
```

```python
# Mean
print("Mean")
for test_column in TEST_COLUMNS:
  test_arr = df[test_column].replace([np.inf, -np.inf], np.nan).dropna().to_numpy()
  # test_arr = test_arr[~np.isnan(test_arr)]

  print(test_column+":", np.mean(test_arr))
```

```python
# Median
print("Median")
for test_column in TEST_COLUMNS:
  test_arr = df[test_column].replace([np.inf, -np.inf], np.nan).dropna().to_numpy()

  print(test_column+":", np.median(test_arr))
```

```python
# STD
print("STD")
for test_column in TEST_COLUMNS:
  test_arr = df[test_column].replace([np.inf, -np.inf], np.nan).dropna().to_numpy()

  print(test_column+":", np.std(test_arr))
```

```python
# T-Test
print("T-Test")
from scipy import stats

t_stat = stats.ttest_ind(a=df[TEST_COLUMNS[0]].replace([np.inf, -np.inf], np.nan).dropna().to_numpy(),b=df[TEST_COLUMNS[1]].replace([np.inf, -np.inf], np.nan).dropna().to_numpy())
print(t_stat)
t_test_stat = t_stat.statistic
print("t_test_stat:",t_test_stat)
```
